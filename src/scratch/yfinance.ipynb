```python
import yfinance as yf
```

```python
def get_ticker(symbol):

    return yf.Ticker(symbol)
    
ticker = get_ticker("KPITTECH.NS")

# get all stock info
ticker.info
```

```python
# get historical market data
hist = ticker.history(period="1mo")
hist
```
